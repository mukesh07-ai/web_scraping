```python
table = a.find_all('table')[1]
```

```python
print(table)
```

```python
c_titles = table.find_all('th')
```

```python
c_titles
```

```python
c_t_t = [title.text.strip() for title in c_titles]
```

```python
c_t_t
```

```python
df = pd.DataFrame(columns = c_t_t)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
